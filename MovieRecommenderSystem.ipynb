```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('/content/drive/MyDrive/dataset/tmdb_5000_movies.csv')
credits=pd.read_csv('/content/drive/MyDrive/dataset/tmdb_5000_credits.csv')
```

```python
#Which columns to keep
#genre,id,keywords,title,overview,cast,crew

movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]



```

```python
import ast

def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
def convert3(obj):
  counter=0
  L=[]
  for i in ast.literal_eval(obj):
    if counter!=3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['cast']+movies['crew']
```

```python
print(movies['tags'][0])
```

```python
movies.head(1)
df=movies[['movie_id','title','tags']]
```

```python
df['tags']=df['tags'].apply(lambda x:" ".join(x))
```

```python
df.head()
```

```python
df['tags']=df['tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000, stop_words='english')
cv.fit_transform(df['tags']).toarray()
```

```python
vectors=cv.fit_transform(df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
#stemming

from nltk.stem.porter import PorterStemmer

ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000, stop_words='english')
cv.fit_transform(df['tags']).toarray()
vectors=cv.fit_transform(df['tags']).toarray()
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
def recommend(movie):
  movie_index=df[df['title']==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)),key=lambda x:x[1],reverse=True)[1:6]
  for i in movies_list:
    print(df.iloc[i[0]].title)
```

```python
recommend('Batman')
```
